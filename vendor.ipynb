```python
vendor_sales_summary = pd.read_sql_query("""
WITH FreightSummary AS (
    SELECT 
        vi."vendornumber" AS "VendorNumber",
        SUM(vi."freight") AS "FreightCost"
    FROM vendor_invoice vi
    GROUP BY vi."vendornumber"
),
PurchaseSummary AS (
    SELECT 
        p."VendorNumber",
        p."VendorName", 
        p."Brand",
        p."Description",
        p."Pur
        chasePrice",
        pp."volume",
        pp."price" AS "ActualPrice",
        SUM(p."Quantity") AS "TotalPurchaseQuantity", 
        SUM(p."Dollars") AS "TotalPurchaseDollars"
    FROM purchases p
    JOIN purchase_prices2 pp
        ON p."Brand" = pp."brand"
    WHERE p."PurchasePrice" > 0
    GROUP BY p."VendorNumber", p."VendorName", p."Brand", p."Description",
             p."PurchasePrice", pp."volume", pp."price"
),
SalesSummary AS (
    SELECT 
       s."VendorNo" AS "VendorNumber",
       s."Brand",
       SUM(s."SalesQuantity") AS "TotalSalesQuantity",
       SUM(s."SalesDollars") AS "TotalSalesDollars",
       SUM(s."SalesPrice") AS "TotalSalesPrice",
       SUM(s."ExciseTax") AS "TotalExciseTax"
    FROM sales s
    GROUP BY s."VendorNo", s."Brand"
)
SELECT 
    ps."VendorNumber",
    ps."VendorName",
    ps."Brand",
    ps."Description",
    ps."PurchasePrice",
    ps."volume",
    ps."ActualPrice",
    ps."TotalPurchaseQuantity",
    ps."TotalPurchaseDollars",
    ss."TotalSalesQuantity",
    ss."TotalSalesDollars",
    ss."TotalSalesPrice",
    ss."TotalExciseTax",
    fs."FreightCost"
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps."VendorNumber" = ss."VendorNumber"
    AND ps."Brand" = ss."Brand"
LEFT JOIN FreightSummary fs
    ON ps."VendorNumber" = fs."VendorNumber"
ORDER BY ps."TotalPurchaseDollars";
""", engine)
```

```python
from sqlalchemy.dialects.postgresql import insert
from sqlalchemy import Table, MetaData

t = Table("vendor_sales_summary1", MetaData(), autoload_with=engine)
ins, upd = 0, 0

with engine.begin() as conn:
    for _, r in vendor_sales_summary.iterrows():
        stmt = insert(t).values(r.to_dict()).on_conflict_do_update(
            index_elements=["VendorNumber", "Brand"],
            set_={c: r[c] for c in r.index if c not in ["VendorNumber", "Brand"]}
        )
        result = conn.execute(stmt)
        if result.rowcount == 1:
            ins += 1
        else:
            upd += 1

print(f"Inserted: {ins}, Updated: {upd}")
```

```python
import pandas as pd
import os
import logging
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

os.makedirs("logs", exist_ok=True)
# Setup logging
logging.basicConfig(
    filename="logs/get_vendor_summary.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

# Setup logging (file + console)
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler("logs/get_vendor_summary.log", mode="a"),
        logging.StreamHandler()  # also prints to console/notebook
    ]
)
# Database connection (URL encoding for password with @)
url = URL.create(
    drivername="postgresql+psycopg2",
    username="postgres",
    password="Aditi@123",   # safe here, SQLAlchemy handles it
    host="localhost",
    port=5432,
    database="postgres"
)

engine = create_engine(url)

def create_vendor_summary(engine):
    query = """
    WITH FreightSummary AS (
        SELECT 
            vi."vendornumber" AS "VendorNumber",
            SUM(vi."freight") AS "FreightCost"
        FROM vendor_invoice vi
        GROUP BY vi."vendornumber"
    ),
    PurchaseSummary AS (
        SELECT 
            p."VendorNumber",
            p."VendorName", 
            p."Brand",
            p."Description",
            p."PurchasePrice",
            pp."volume",
            pp."price" AS "ActualPrice",
            SUM(p."Quantity") AS "TotalPurchaseQuantity", 
            SUM(p."Dollars") AS "TotalPurchaseDollars"
        FROM purchases p
        JOIN purchase_prices2 pp
            ON p."Brand" = pp."brand"
        WHERE p."PurchasePrice" > 0
        GROUP BY p."VendorNumber", p."VendorName", p."Brand", p."Description",
                 p."PurchasePrice", pp."volume", pp."price"
    ),
    SalesSummary AS (
        SELECT 
           s."VendorNo" AS "VendorNumber",
           s."Brand",
           SUM(s."SalesQuantity") AS "TotalSalesQuantity",
           SUM(s."SalesDollars") AS "TotalSalesDollars",
           SUM(s."SalesPrice") AS "TotalSalesPrice",
           SUM(s."ExciseTax") AS "TotalExciseTax"
        FROM sales s
        GROUP BY s."VendorNo", s."Brand"
    )
    SELECT 
        ps."VendorNumber",
        ps."VendorName",
        ps."Brand",
        ps."Description",
        ps."PurchasePrice",
        ps."volume",
        ps."ActualPrice",
        ps."TotalPurchaseQuantity",
        ps."TotalPurchaseDollars",
        ss."TotalSalesQuantity",
        ss."TotalSalesDollars",
        ss."TotalSalesPrice",
        ss."TotalExciseTax",
        fs."FreightCost"
    FROM PurchaseSummary ps
    LEFT JOIN SalesSummary ss
        ON ps."VendorNumber" = ss."VendorNumber"
        AND ps."Brand" = ss."Brand"
    LEFT JOIN FreightSummary fs
        ON ps."VendorNumber" = fs."VendorNumber"
    ORDER BY ps."TotalPurchaseDollars";
    """
    return pd.read_sql_query(query, engine)

def clean_data(df):
    df['volume'] = df['volume'].astype(float)
    df.fillna(0, inplace=True)

    df['VendorNumber'] = df['VendorNumber'].astype(str).str.strip()
    df['Description'] = df['Description'].astype(str).str.strip()

    # New columns
    vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
    vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars']) * 100
    vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity'] / vendor_sales_summary['TotalPurchaseQuantity']
    vendor_sales_summary['SalesToPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars'] / vendor_sales_summary['TotalPurchaseDollars']

    return df

def ingest_df(df, table_name, engine):
    df.to_sql(table_name, engine, if_exists="replace", index=False)

if __name__ == "__main__":
    logging.info("Creating Vendor Summary Table") 
    summary_df = create_vendor_summary(engine)
    logging.info(summary_df.head()) 
    
    logging.info("Cleaning data") 
    clean_df = clean_data(summary_df) 
    logging.info(clean_df.head()) 
    
    logging.info("Ingesting data into vendor_sales_summary")
    ingest_df(clean_df, "vendor_sales_summary", engine) 
    logging.info("Completed")
```
